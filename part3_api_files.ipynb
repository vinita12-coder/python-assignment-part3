```python
!pip install requests

import requests

BASE_URL = "https://dummyjson.com/products"

# Step 1 — Fetch and Display 20 Products
# ========================================

response = requests.get(f"{BASE_URL}?limit=20", timeout=5)
data     = response.json()
products = data["products"]

# Print Header of  table
print(f"{'ID':<4} | {'Title':<30} | {'Category':<14} | {'Price':<8} | Rating")
print("-" * 70)
for product in products:
  print(f"{product['id']:<4} | {product['title']:<30} | {product['category']:<15} | ${product['price']:<8} | {product['rating']}")


# Step 2 — Filtering rating >= 4.5, sorting by price descending
# ==========================================================

filtered = [p for p in products if p["rating"] >= 4.5]
filtered.sort(key=lambda x: x["price"], reverse=True)

print("\n--- Products with Rating >= 4.5 (sorted by price) ---")
for p in filtered:
    print(f"{p['title']:<30} | ${p['price']} | Rating: {p['rating']}")


# Step 3 — Fetching laptops category
# ==================================
laptop_response = requests.get(f"{BASE_URL}/category/laptops", timeout=5)
laptops         = laptop_response.json()["products"]

print("\n--- Laptops ---")
for laptop in laptops:
    print(f"{laptop['title']} — ${laptop['price']}")


# Step 4 — POST request
# ========================================

new_product = {
    "title": "My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

post_response = requests.post(f"{BASE_URL}/add", json=new_product, timeout=5)
print("\n--- POST Response ---")
print(post_response.json())
```

# ***Task3: Exceptional Handling***

```python
import requests

# Part A — Safe Divide Function
# =================================

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print("Part A - Safe Divide:")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# Part B — Guarded File Reader
# ================================

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    finally:
        print("File operation attempt complete.")

print("\nPart B - Safe File Reader:")
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))


# Part C — Robust API Calls
# ============================

BASE_URL = "https://dummyjson.com/products"

def fetch_products():
    try:
        response = requests.get(f"{BASE_URL}?limit=20", timeout=5)
        return response.json()["products"]
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")


def fetch_laptops():
    try:
        response = requests.get(f"{BASE_URL}/category/laptops", timeout=5)
        return response.json()["products"]
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")


def post_product(product):
    try:
        response = requests.post(f"{BASE_URL}/add", json=product, timeout=5)
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")

print("\n PartC: API Calls with Exception Handling:")
products = fetch_products()
if products:
  print(f"Fetched {len(products)} products successfully.")


# Part D — Input Validation Loop
# ==================================

print("\n PartD - Product Lookup:")

while True:
    user_input = input("\nEnter a product ID (1–100), or 'quit' to exit: ")

    # Exit condition
    if user_input.lower() == "quit":
        print("Exiting.")
        break

    # Validate: must be an integer
    if not user_input.isdigit():
        print("⚠ Invalid input: please enter a number.")
        continue

    product_id = int(user_input)

    # Validate: must be in range 1-100
    if product_id < 1 or product_id > 100:
        print("⚠ Invalid input: ID must be between 1 and 100.")
        continue

    # Valid input — make API call
    try:
        response = requests.get(f"{BASE_URL}/{product_id}", timeout=5)

        if response.status_code == 404:
            print("Product not found.")
        elif response.status_code == 200:
            product = response.json()
            print(f"Title : {product['title']}")
            print(f"Price : ${product['price']}")
        else:
            print(f"Unexpected status code: {response.status_code}")

    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    except Exception as e:
        print(f"Unexpected error: {e}")
```

# ***Task4: Logging to File***

```python
import requests
from datetime import datetime

# Logger Function — writes timestamped entry to error_log.txt
# ============================================================

def log_error(location, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log_entry = f"[{timestamp}] ERROR in {location}: {error_type} — {message}\n"

    with open("error_log.txt", "a", encoding="utf-8") as f:
        f.write(log_entry)

    print(f"Logged: {log_entry.strip()}")


# Trigger 1 — ConnectionError
# ==============================

try:
    response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError", str(e))
except requests.exceptions.Timeout:
    log_error("fetch_products", "Timeout", "Request timed out")
except Exception as e:
    log_error("fetch_products", "UnexpectedError", str(e))


# Trigger 2 — HTTP 404
# ========================

try:
    response = requests.get("https://dummyjson.com/products/999", timeout=5)

    # 404 is NOT a python exception — check status code manually
    if response.status_code != 200:
        log_error(
            "lookup_product",
            "HTTPError",
            f"{response.status_code} Not Found for product ID 999"
        )
    else:
        print("Product found:", response.json()["title"])

except requests.exceptions.ConnectionError as e:
    log_error("lookup_product", "ConnectionError", str(e))
except requests.exceptions.Timeout:
    log_error("lookup_product", "Timeout", "Request timed out")
except Exception as e:
    log_error("lookup_product", "UnexpectedError", str(e))


# Print full error_log.txt
# ============================

print("\n--- error_log.txt contents ---")
with open("error_log.txt", "r", encoding="utf-8") as f:
    print(f.read())
```
